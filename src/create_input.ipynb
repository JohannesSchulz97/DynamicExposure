```python
for seed in seeds: 
    network = Network(N=N, K_aff=K_aff, seed=seed, dirichlet=dirichlet, prob_ratio=prob_ratio)
    u = network.get_u()
    for T in Ts: 
        A0 = network.generate_A0(T=T, avg_degree=avg_degree_A0, symmetric=symmetric)
        w = network.get_w()
        for K_exp in K_exps: 
            A1 = network.apply_exposure(A0, K_exp)
            mu, Z = network.get_mu(), network.get_Z()
            file = f"{seed}_{T}_{K_exp}"
            write_data(folder, file, A1)
            np.savez(folder+"params_" + file, mu=mu, Z=Z, u=u, w=w)
```
